```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# Load Data set
athletes = pd.read_csv("C:/Users/Owner/Desktop/Data Analyst/Python Data Analysis/Olympics Data/athlete_events.csv")
region = pd.read_csv("C:\\Users\\Owner\\Desktop\\Data Analyst\\Python Data Analysis\\Olympics Data\\noc_regions.csv")
```

```python
# Top countries participating

top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for the top 10 countries

plt.figure(figsize=(12,6))
#plt.xticks(rotation=20)
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette = 'Set2');
```

```python
#Age Distribution of the partipipants

plt.figure(figsize=(12,6))
plt.title("Age Distribution of the Athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color='green', edgecolor = 'white');
```

```python
winter_sports = athletes_df[athletes_df.Season == "Winter"].Sport.unique()
winter_sports
```

```python
# Summer olympics sports

summer_sports = athletes_df[athletes_df.Season == "Summer"].Sport.unique()
summer_sports
```

```python
# Male and Female participants

gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie plot for male and female participants

plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle=150, shadow =True);
```

```python
# Total medals
Medal_counts = athletes_df.Medal.value_counts()
Medal_counts
```

```python
#Pie chart showing Medal counts

plt.figure(figsize=(12,6))
plt.title('Medal Distribution')
plt.pie(Medal_counts, labels = Medal_counts.index, autopct = '%1.1f%%', startangle=150, shadow =True);
```

```python
# Total number of female athletes in each olympics.

female_participants  = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
#Bar chart for Women Participation in Summer Olympics
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlympics, palette="Spectral")
plt.title('Women Participation in Summer Olympics')
```

```python
#Line graph showing female athletes overtime
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:, 'F'].plot()
plt.title('Women participation in Summer Olympics over time')
```

```python
# Gold medal Athletes

gold_medals = athletes_df[(athletes_df.Medal == "Gold")]
gold_medals.head()
```

```python
# take only the values that are different from NaN.

gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
# Gold beyond 60 years of age

gold_medals["ID"][gold_medals['Age']> 60].count()
```

```python
# sporting event for Gold  beyond 60 

sporting_event = gold_medals['Sport'][gold_medals["Age"]>60]
sporting_event
```

```python
# Plot for sporting_event

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold Medals won by Athletes over 60 years")
```

```python
# Gold medals from each country

gold_medals.region.value_counts().reset_index(name='Medal').head(5)
```

```python
total_gold_medals = gold_medals.region.value_counts().reset_index(name="Medal").head(5)
g = sns.catplot(x="index", y = "Medal", data=total_gold_medals, 
               height=5, kind = "bar", palette = "tab10")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title("Top 5 Countries with highest Gold Medals")
```

```python
# Rio olympics

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == "Gold")].Team

team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Top 20 Countries with Highest Gold Medals in 2016')
```
